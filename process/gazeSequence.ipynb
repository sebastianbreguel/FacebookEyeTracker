```python
import pandas as pd
import json
from datetime import datetime, timedelta
import os
import matplotlib.pyplot as plt

# Load the gaze data


def moving_average(data, window_size):
    return data.rolling(window=window_size).mean()


gaze_data = pd.read_csv(
    "eye_trackers/my_gaze_bea_1.csv"
)  # Replace with the actual path to your data file
gaze_data = gaze_data[gaze_data["isnan"] == False]


for i in range(1, 15, 2):
    window_size = i
    gaze_data_avg = gaze_data[["x", "y"]].apply(
        lambda x: moving_average(x, window_size)
    )

    # Load the background image
    img = plt.imread("images/1.png")

    # Plot the image
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.imshow(img)

    # Plot the averaged gaze points with adjusted size and transparency
    ax.scatter(gaze_data_avg["x"], gaze_data_avg["y"], s=50, c="purple", alpha=0.3)

    # Connect the points with lines to show the sequence
    ax.plot(gaze_data_avg["x"], gaze_data_avg["y"], color="purple", alpha=0.5)

    # Optionally, label the first few points to illustrate the gaze sequence
    for i in range(
        min(10, len(gaze_data_avg))
    ):  # Label the first 10 points or less if data has fewer points
        ax.text(
            gaze_data_avg["x"][i],
            gaze_data_avg["y"][i],
            str(i + 1),
            fontsize=8,
            color="white",
            ha="center",
            va="center",
        )

plt.show()
```
